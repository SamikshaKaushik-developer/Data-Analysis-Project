```python
import pandas as pd

campaign=pd.read_csv("/content/campaign_performance.csv")
campaign.head(5)
```

```python
# Check the current data types
print(campaign.dtypes)
```

```python
# Remove duplicate rows
campaign= campaign.drop_duplicates()
print(f"Number of rows after removing duplicates: {len(campaign)}")
```

```python
unique_values = campaign['platform'].unique()
print(f"The distinct platform are: {unique_values}")
```

```python
campaign['platform'] = campaign['platform'].str.replace('Instagram', 'Instagram')
campaign['platform'] = campaign['platform'].str.replace('insta', 'Instagram')

campaign['platform'] = campaign['platform'].str.replace('Instagram Ads', 'Instagram')
campaign['platform'] = campaign['platform'].str.replace('YouTube', 'YouTube')
campaign['platform'] = campaign['platform'].str.replace('YT', 'YouTube')
campaign['platform'] = campaign['platform'].str.replace('YouTube Ads', 'YouTube')

campaign['platform'] = campaign['platform'].str.replace('LI', 'LinkedIn')
campaign['platform'] = campaign['platform'].str.replace('LinkedIn Ads', 'LinkedIn')
unique_values = campaign['platform'].unique()
print(f"The distinct platform are: {unique_values}")
```

```python
campaign_sorted = campaign.sort_values(by=['ad_spend', 'impressions' , 'clicks'], ascending=[True, True, True])
print(campaign_sorted.head(10))
```

```python
value_counts = campaign['platform'].value_counts()
print("Count of each distinct product:")
print(value_counts)
```

```python
campaign.to_csv('final_campaign_dataset.csv', index=False)
```

```python
enrollments=pd.read_csv("/content/student_enrollments .csv")
enrollments.head(5)
```

```python
enrollments= enrollments.dropna()
print(f"Number of rows after removing null values: {len(enrollments)}")
```

```python
enrollments =enrollments.drop_duplicates()
print(f"Number of rows after removing duplicates: {len(enrollments)}")
```

```python
# Check the current data types
print(enrollments.dtypes)
```

```python
enrollments['course_name'] = enrollments['course_name'].str.title()
enrollments.head()
```

```python
value_counts_enroll = enrollments['course_name'].value_counts()
print("Count of each distinct product:")
print(value_counts_enroll)
```

```python
enrollments['signup_date'] = pd.to_datetime(enrollments['signup_date'])

# Extract month and day of week
enrollments['month'] = enrollments['signup_date'].dt.month
enrollments['year'] = enrollments['signup_date'].dt.year

enrollments['day_of_week'] = enrollments['signup_date'].dt.day_name()

print(enrollments.head())
```

```python
enrollments_sorted = enrollments.sort_values(by=['year', 'month'], ascending=[True, True])
print(enrollments.head())
```

```python
enrollments_sorted.to_csv('final_enrollment_dataset.csv', index=False)
```
